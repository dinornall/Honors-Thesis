```python
# --- Main symbols
theta, psi = dynamicsymbols('theta, psi')
theta_s = sp.symbols('theta_s')
#omegax, omegay, omegaz = dynamicsymbols('omega_x, omega_y, omega_z')
#Mx, My, Mz = dynamicsymbols('M_x, M_y, M_z') # Torques on Top
#theta0, phi0, Ktheta, Kphi = sp.symbols('theta_0, phi_0, K_theta, K_phi') # Restoring forces
#r = sp.symbols('r')
#m_n = sp.symbols('m_n')
#JGn_xx , JGn_yy , JGn_zz = sp.symbols('JGn_xx , JGn_yy ,JGn_zz ')
#x_NGn, r_NGn_z = sp.symbols('r_NGn_x, r_NGn_z')

m_b = sp.symbols('m_n')
#JGn_xx , JGn_yy , JGn_zz = sp.symbols('JGn_xx , JGn_yy ,JGn_zz ')
#r_NGn_x, r_NGn_z = sp.symbols('r_NGn_x, r_NGn_z')

x_NR, z_NR = sp.symbols('x_NR, z_NR')

x_GB, y_GB = sp.symbols('x_GB, y_GB')

gravity = sp.symbols('g') # Gravity
time = dynamicsymbols._t
#Jxx, Jyy, Jzz, JO, JG = sp.symbols('Jxx, Jyy, Jzz, J_O, J_G')
```

```python
simpleSubs = [(theta_s,0) , (x_GB,0), (y_GB,0)]
```

```python
# --- Bodies and connections
ref = YAMSInertialBody('E') 
nac = YAMSRigidBody('N', J_form='diag') #, mass=m_n, J_G=(JGn_xx , JGn_yy , JGn_zz), rho_G=(r_NGn_x, 0, r_NGn_z), J_at_Origin=True) # Create a body named "B"
bld = YAMSRigidBody('B', J_form='diag')
ref.connectTo(nac, type='Free', rel_pos=(0, 0, 0), rot_type='Body', rot_amounts=(0,  theta, 0), rot_order='XYZ')
nac.connectTo(bld, type='Joint', rel_pos=(x_NR, 0, z_NR), rot_type='Body', rot_amounts=(theta_s, psi, 0), rot_order='YXZ')
```

```python
nac.frame.x.express(ref.frame)
bld.frame.x.express(ref.frame).subs(simpleSubs)
bld.frame.z.express(ref.frame).subs(simpleSubs)
```

```python
bld.kinetic_energy(ref.frame).subs(simpleSubs)
```
